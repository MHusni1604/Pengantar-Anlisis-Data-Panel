```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.regression.mixed_linear_model import MixedLM
from scipy import stats
```

## Load Data


```python
df = pd.read_excel("https://github.com/MHusni1604/Pengantar-Anlisis-Data-Panel/raw/main/Dataset/Panel%20Dataset.xlsx")
df.set_index(['Provinsi ', 'Tahun'], inplace=True)
df.head()
```

## Statistik Deskriptif

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(10, 6))
sns.pairplot(df, kind='reg')
plt.suptitle('Pairplot of Data', y=1.02)
plt.show()
```

## Analisis Regresi Panel Efek Random

```python
X = df[['PDRB', 'Imunisasi Dasar', 'Presentasi Miskin', 'Rata Lama Pendidikan',]]
y = df['Prevalensi Balita Gizi Kurang (0-59 Bulan)']
```

```python
X = sm.add_constant(X)
model = MixedLM(endog=y, exog=X, groups=df.index.get_level_values("Provinsi "))
results = model.fit()
print(results.summary())
```

## Uji Asumsi

```python
# Uji Hausman
model_fixed = sm.OLS(y, X).fit()
llf_mixed = results.llf
llf_ols = model_fixed.llf
lrt_stat = 2 * (llf_mixed - llf_ols)
from scipy.stats import chi2
lrt_pval = chi2.sf(lrt_stat, df=1)

print(f"LRT p-value: {lrt_pval}")
```

```python
# Uji Jarque-Bera
from statsmodels.stats.stattools import jarque_bera
jb_test_stat, jb_p_value, _, _ = jarque_bera(results.resid)
print(f"Jarque-Bera Test Statistic: {jb_test_stat}, p-value: {jb_p_value}")
```

```python
# Uji Breusch-Pagan
from statsmodels.stats.diagnostic import het_breuschpagan
bp_test_stat, bp_p_value, _, _ = het_breuschpagan(results.resid, X)
print(f"Breusch-Pagan Test Statistic: {bp_test_stat}, p-value: {bp_p_value}")
```

```python
# Uji Durbin-Watson
from statsmodels.stats.stattools import durbin_watson
dw_test_stat = durbin_watson(results.resid)
print(f"Durbin-Watson Test Statistic: {dw_test_stat}")
```

```python
# Multicolinearity (VIF)
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif_data)
```
